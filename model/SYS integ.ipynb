```python
class_names = [
    "Mildiou", "oidium", "alternariose", "sunburn",
    "helicoverpa", "fruit_saine", "feuil_saine", "excesnit",
    "blossom_end_rot", "tuta", "bacterial", "mited",
    "virosis", "black mold"
]
```

```python
import tensorflow as tf
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import numpy as np

# 1. Charger le mod√®le
model = load_model('D:/PFE RANIA/models/alternaria mite/1')

# 2. Pr√©parer une image pour la pr√©diction
def preprocess_image(img_path, target_size=(224, 224)):
    img = image.load_img(img_path, target_size=target_size)  # adapte la taille au mod√®le
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)  # ajouter la dimension batch
    img_array /= 255.0  # normaliser si n√©cessaire
    return img_array

# 3. Charger et pr√©traiter l'image
img_path = 'C:/Users/Rania-Bou/Desktop/PFE RANIA/TOM2024/CATEGORY A/CATA-English/tomato_diseases/alternaria_mite_d/779509792_1660212901675.jpg'
img_prepared = preprocess_image(img_path, target_size=(224, 224))  # adapte la taille

# 4. Faire la pr√©diction
preds = model.predict(img_prepared)

# 5. Interpr√©ter la pr√©diction selon ton mod√®le
print("Pr√©diction :", preds)
```
